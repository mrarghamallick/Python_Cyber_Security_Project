```python
import requests
import json
from datetime import datetime
 
api_key = "ea23220933d0f62ef676a4f77991ee84"
loc = input("Enter the name of the CITY: ")
 
url = "https://api.openweathermap.org/data/2.5/weather?q="+loc+"&appid="+api_key
 
res = requests.get(url)
data = res.json()
 
temp = ((data['main']['temp']) - 273.15)
real_feel = ((data['main']['feels_like']) - 273.15)
des = data['weather'][0]['description']
hmdt = data['main']['humidity']
pres = data['main']['pressure']
wind_spd = data['wind']['speed']
date_time = datetime.now().strftime("%d %b %Y at %I:%M:%S %p GMT")
 
print ("---------------------------------------------------------------")
print ("Weather Report for - {}  on {}".format(loc.upper(), date_time))
print ("---------------------------------------------------------------")
print ("---------------------------------------------------------------")
print ("Current temperature is ----------- : {:.2f} deg C".format(temp))
print ("Current Real-Feel temperature is - : {:.2f} deg C".format(real_feel))
print ("Current weather Condition -------- :",des)
print ("Current Humidity ----------------- :",hmdt, '%')
print ("Current Pressure ----------------- :",pres,'mili bar')
print ("Current Wind-speed --------------- :",wind_spd ,'kmph')
print ("---------------------------------------------------------------")
```
